```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("Datetime").getOrCreate()
```

```python
apple = spark.read.csv("AAPL_data.csv", inferSchema= True, header = True)
```

```python
 apple.show()
```

```python
from pyspark.sql.functions import (dayofmonth, dayofweek, month, year,
                                   weekofyear,dayofyear, hour, format_number, date_format)
```

```python
apple.select(month("date")).show()
apple.select(year("date")).show()
```

```python
apple.select(year("date")).show()
```

```python
apple.select(hour(apple["date"])).show()
```

```python
new_df= apple.withColumn("Year", year(apple["date"]))
```

```python
new_df.show()
```

```python
new_df.select("Year").distinct().show()
```

```python
new = new_df.groupBy("Year").mean().select(["Year", "avg(close)"])
```

```python
new.select(["Year", format_number("avg(close)", 2).alias("Average Closing Price")]).orderBy("Year").show()
```
